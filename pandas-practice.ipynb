# #merge()

```python
df1 = pd.DataFrame({'Roll NO.':[1,2,3,4,5], 'Physics':[34,64,34,89,12]})
df1
```

```python
df2 = pd.DataFrame({'Roll NO.':[1,2,3,4,5], 'Chemistry':[78,33,39,81,90]})
df2
```

```python
pd.merge(df1,df2, on ='Roll NO.')
```

```python
pd.merge(df2,df1, on ='Roll NO.')
```

```python
pd.merge(df1,df2)
```

```python
df3 = pd.DataFrame({'Roll NO.':[1,2,3,6,7], 'Physics':[34,67,34,89,12]})
df4 = pd.DataFrame({'Roll NO.':[1,2,3,8,9], 'Chemistry':[34,67,34,89,12]})
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4, how = 'left')
```

```python
pd.merge(df3,df4, how = 'right')
```

```python
pd.merge(df3,df4, how = 'outer')
```

# #append() Function****

```python
df1 = pd.DataFrame({'Roll NO.':[1,2,3,4,5], 'Maths':[45,78,45,90,66], 'Physics':[33,67,12,90,44]})
df2 = pd.DataFrame({'Roll NO.':[6,7,8,9,10], 'Maths':[78,75,45,90,69], 'Physics':[23,67,88,0,98]})
print(df1)
print(df2)
```

```python
df1.append(df2)
```

```python
df2.append(df1)
```

```python
df1.append(df2, ignore_index = True)
```

```python
df1.append(df2, ignore_index = True, sort = True)
```

```python
df1 = pd.DataFrame({'Roll NO.':[1,2,3,4,5], 'Chemistry':[45,78,45,90,66], 'Physics':[33,67,12,90,44]})
df2 = pd.DataFrame({'Roll NO.':[6,7,8,9,10], 'Maths':[78,75,45,90,69], 'Physics':[23,67,88,0,98]})
print(df1)
print(df2)
```
